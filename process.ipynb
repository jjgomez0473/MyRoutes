```python
import streamlit as st
import pandas as pd

# Genera tu dataset en Pandas (aquí hay un ejemplo simple)
data = {
    'Nombre': ['Juan', 'María', 'Pedro'],
    'Edad': [25, 30, 35],
    'Ciudad': ['México', 'Madrid', 'Buenos Aires']
}
df = pd.DataFrame(data)

# Define la función para descargar el archivo Excel
def descargar_excel(df):
    excel_file = df.to_excel(index=False)
    with open("dataset.xlsx", "wb") as excel_writer:
        excel_writer.write(excel_file.getbuffer())
    return excel_file

# Crear la aplicación de Streamlit
st.title('Descargar Dataset como Excel')

# Botón para descargar el archivo
if st.button('Descargar Excel'):
    st.write('Descargando dataset como archivo Excel...')
    descargar_excel(df)
    st.success('¡Descarga exitosa! El archivo "dataset.xlsx" se ha descargado.')

```
